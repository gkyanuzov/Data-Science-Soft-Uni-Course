```python
weather_data = pd.read_csv('weather.csv')
```

```python
weather_data
```

```python
weather_data.element.unique()
```

```python
#melt - pivot_table - obratni operacii
weather_data = weather_data.melt(id_vars = ['id', 'year', 'month', 'element'], var_name = 'day')
```

```python
#removing the "d" and transforming days into int
weather_data.day = weather_data.day.str.slice(1).astype(int) #prezapisvame promenenata kolona

#weather_data.day = weather_data['day']
```

```python
weather_data = weather_data.pivot_table(index  = ['id', 'year', 'month', 'day'], columns = ['element'],
                        values = 'value')
#pivot_table automatically removes NAs if not specified otherwise
```

```python
weather_data.index
```

```python
weather_data = weather_data.reset_index()
```
